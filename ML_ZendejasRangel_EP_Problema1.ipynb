```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Hacemos lectura del dataset con pandas.
lec=pd.read_csv('dataset1.csv',names=['a','b'])
```

```python
#Usamos la función predeterminada de pandas para graficar el scatter.
lec.plot.scatter(['a'],['b'])
```

```python
#Utilizamos las funciones predeterminadas para Kmeans, para colocar 5 centros, luego ajustamos la información con la función
# .fit. Finalmente predecimos los posibles k-centros
kmeans = KMeans(n_clusters=5)
kmeans.fit(lec)
y_kmeans = kmeans.predict(lec)
```

```python
#Graficamos los puntos y los centros.
plt.scatter(lec['a'], lec['b'], c=y_kmeans, s=50, cmap='spring')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha=0.8)
```
