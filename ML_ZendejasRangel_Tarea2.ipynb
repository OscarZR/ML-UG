```python
import numpy as np
import sympy
```

```python
#Usando numpy
A=np.array([[2,1,1],[0,1,-1],[0,2,4]])
B=np.array([[4,0,1],[-2,1,0],[-2,0,1]])
C,E=np.linalg.eig(A)
D,F=np.linalg.eig(B)
print(C,'\n',E)
print('\n',D,'\n',F)
```

```python
#Usando sympy.
a=sympy.Matrix([[2,1,1],[0,1,-1],[0,2,4]])
b=sympy.Matrix([[4,0,1],[-2,1,0],[-2,0,1]])
c=a.eigenvects()
d = b.eigenvects()
print(c)
print('\n')
print(d)
```
