```python
#Importamos las librer√≠as necesarias.
%matplotlib inline
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
training_data,labels=make_blobs(n_samples=200,centers=3,cluster_std=0.5)
test_data,labels2=make_blobs(n_samples=10,centers=3,cluster_std=0.5)
```

```python
aux=[];p=[];c=[]
for i in range(3):
    for j in range(200):
        if labels[j]==i:
            aux.append(training_data[j])
    p.append(aux)
    aux=[]
    c.append((np.sum(p[i],axis=0))/len(p[i]))
c=np.array(c)
```

```python
plt.scatter(training_data[:, 0], training_data[:, 1], c=labels, s=50, alpha=0.3, cmap='spring')
plt.scatter(c[:, 0], c[:, 1], c='blue', s=100, alpha=0.9)
```

```python
def distancia(c,v):
    dis=[]
    for n in range(3):
        dis.append(np.sqrt(np.sum((c[n]-v)**2)))
    dis_min=np.where(dis==np.amin(dis))
    return dis,dis_min

M=[];PreLab=[]
for i in test_data:
    aux,aux2=distancia(c,i)
    M.append(aux) 
    PreLab.append(aux2)
```
